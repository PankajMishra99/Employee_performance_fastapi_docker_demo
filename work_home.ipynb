```python
# encode target column in the category.. 
def burnout_risk(data):
    if data=='Low':
        return 1 
    elif data=='Medium':
        return 2 
    else:
        return 3

df_copy['burnout']=df_copy['burnout_risk'].apply(burnout_risk)
df_copy
```

```python
#  encode day_type column 
def day_type(data):
    if data=='Weekday':
        return 1 
    else:
        return 2 

df_copy['day_types']=df_copy['day_type'].apply(day_type)
df_copy
```

```python
from sklearn.feature_selection import RFE,SelectFromModel
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier 
```

```python
input_col=['user_id','day_types','work_hours','screen_time_hours','meetings_count','breaks_taken','after_hours_work','sleep_hours','task_completion_rate',
'burnout_score']
output_col='burnout'
```

```python
x=df_copy[input_col]
x
```

```python
y=df_copy[output_col]
y.unique()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
model=RandomForestClassifier(max_depth=10)
pipline = Pipeline(steps=[
    ('rfe', SelectFromModel(estimator=model)),
    ('model', model)
])

pipline.fit(x_train,y_train)
```

```python
y_pred = pipline.predict(x_test)
y_pred
```

```python
from sklearn.metrics import * 
acc_score=accuracy_score(y_test,y_pred)
acc_score
```

```python
import pickle 
model_name='model.pkl'
with open (model_name,'wb') as f:
    pickle.dump(pipline,f)
```
